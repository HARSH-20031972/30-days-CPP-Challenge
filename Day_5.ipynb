```python
def dimension_is(t):
  if t==1: # 1 is a perfect cube (1*1*1=1)
      return 1
  if t<=0: # 0 is a perfect cube (0*0*0=0), negative numbers are not usually considered
      return 0
  i=1
  result=0
  while i*i*i <= t: # Loop while cube is less than or equal to t
    if i*i*i==t:
      result=1
      break
    i=i+1
  return result
n=int(input("Enter The Number "))
a=list(map(int,input().strip().split()))
perfect_cube=0
for element in a:
  if dimension_is(element)==1:
    perfect_cube+=1
print(perfect_cube)
```

```python
def mode(a,n,value):
  sum=0
  for i in range(0,n):
    sum=sum+(a[i]%value)
  return sum
```
