```python
class Worker:
  def __init__(self,name,hourly_rate):
    self.name=name
    self.hourly_rate=hourly_rate
    self.hours_worked=0
  def work(self,hours):
    self.hours_worked+=hours
    print(f"{self.name} worked for {hours} hours")
  def calculate_pay(self):
    return self.hourly_rate*self.hours_worked
```

```python
class Manager:
  def __init__(self,team_size=0):
    self.team_size=team_size
    self.team_members=[]
  def manage_team(self):
    print(f"Managing Team of {self.team_size} people")
  def add_team_member(self,worker):
    self.team_members.append(worker)
    self.team_size=len(self.team_members)
```

```python
class TeamLead(Worker,Manager):
  def __init__(self,name,hourly_rate,team_size):
    Worker.__init__(self,name,hourly_rate)
    Manager.__init__(self,team_size)
    self.bonus=0
  def set_bonus(self,amount):
    self.bonus=amount
    print(f"{self.name} got a bonus of INR{amount}")
  def calculate_pay(self):
    base_pay=Worker.calculate_pay(self)
    return base_pay+self.bonus
```

```python
lead=TeamLead("Anurag",100, 0)
lead.work(40)
lead.add_team_member("Varun")
lead.add_team_member("Vivek")
lead.manage_team()
lead.set_bonus(10000)
print(f"total pay:- INR{lead.calculate_pay()}")
```

```python
def Brute_force_search(number_list,target):
  for index in range(len(number_list)):
      print(f"Chechking index {index} : is {number_list[index]} equal to target {target}")
      #if current item matches then target we found it
      if number_list[index]==target:
       return index
  return -1
```

```python
my_list=[5,2,9,7,1]
search_target=7
print(f"Searching Brute Force search for {search_target} in {my_list}")
result_index=Brute_force_search(my_list,search_target)
if result_index!=-1:
  print(f"Success ! found {search_target} at index {result_index}")
else:
  print("Result not Found")
```

```python
def find_duplicate(arr:List[int])->int:
    n=len(arr)
    for i in range(n-1):
      for j in range(i+1,n):
        if arr[i]==arr[j]:
          return arr[i]
      return -1
```
